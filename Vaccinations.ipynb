```python
df_temp_people_vaccinated = df_country_vaccinations.groupby(["country", "vaccines"], as_index = False)["people_vaccinated"].max()
```

```python
df_temp_people_vaccinated
```

```python
df_vaccine_people_vaccinated = df_temp_people_vaccinated.groupby(["vaccines"], as_index = False)["people_vaccinated"].sum()
```

```python
df_vaccine_people_vaccinated
```

```python
vaccine_people_vaccinated = dict()
for vaccine in df_vaccine_people_vaccinated["vaccines"]: 
    if "," not in vaccine: 
        vaccine_people_vaccinated[vaccine] = df_vaccine_people_vaccinated.loc[df_vaccine_people_vaccinated["vaccines"] == vaccine, "people_vaccinated"].iloc[0]
    else: 
        vaccine_list = vaccine.split(",")
        for sub_vaccine in vaccine_list: 
            sub_vaccine = sub_vaccine.strip()
            if sub_vaccine in vaccine_people_vaccinated.keys(): 
                vaccine_people_vaccinated[sub_vaccine] = vaccine_people_vaccinated[sub_vaccine] + df_vaccine_people_vaccinated.loc[df_vaccine_people_vaccinated["vaccines"] == vaccine, "people_vaccinated"].iloc[0]
            else: 
                vaccine_people_vaccinated[sub_vaccine] = df_vaccine_people_vaccinated.loc[df_vaccine_people_vaccinated["vaccines"] == vaccine, "people_vaccinated"].iloc[0]
               
        
```

```python
vaccine_people_vaccinated
```

```python
vaccine_pv_col1 = list(vaccine_people_vaccinated.keys())
vaccine_pv_col2 = list(vaccine_people_vaccinated.values())
```

```python
data_vaccine_pv = {"vaccines": vaccine_pv_col1, "people_vaccinated": vaccine_pv_col2}
df_total_vaccine_pv = pd.DataFrame.from_dict(data_vaccine_pv)
```

# Infections WorldWide

```python
df_infections_ww_ds = pd.read_csv("Infections WorldWide.csv")
```

```python
df_infections_ww_ds
```

```python
df_infections_ww_ds.loc[df_infections_ww_ds["Country,Other"] == "India", "Total Cases"]
```

```python
for country in country_list: 
    if country not in list(df_infections_ww_ds["Country,Other"]): 
        print(country)
```

```python
data_country_infections = {"Country" : country_list}
df_country_infections = pd.DataFrame(data_country_infections)
```

```python
name_equivalents = {"Cote d'Ivoire" : "Ivory Coast", "South Korea" : "S. Korea", "United Arab Emirates" : "UAE", "United Kingdom" : "UK",
                   "United States" : "USA", "Turks and Caicos Islands" : "Turks and Caicos"}
```

```python
name_equivalents_old = list(name_equivalents.keys()) 
name_equivalents_new = list(name_equivalents.values())
```

```python
df_country_infections["Country"] = df_country_infections["Country"].replace(name_equivalents_old, name_equivalents_new)
```

```python
country_infections = {"England" : 3822032, "Guernsey" : 821, "Jersey" : 3219, "Northern Cyprus" : 49988, "Northern Ireland" : 118146, 
                     "Saint Vincent and the Grenadines" : 1791, "Scotland" : 222281, "Wales" : 210125}
```

```python
for country in df_country_infections["Country"]: 
    if country in country_infections.keys(): 
        df_country_infections.loc[df_country_infections["Country"] == country, "Total Cases"] = country_infections[country]
    else: 
        try: 
            df_country_infections.loc[df_country_infections["Country"] == country, "Total Cases"] = df_infections_ww_ds.loc[df_infections_ww_ds["Country,Other"] == country, "Total Cases"].iloc[0]
        except: 
            pass
```

```python
df_country_infections.dropna(subset = ["Total Cases"])
```
