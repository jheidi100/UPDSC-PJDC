```python
from textblob import TextBlob
import pandas as pd
import re
```

```python
df = pd.read_csv("vaccination_all_tweets.csv")
```

```python
df
```

```python
df.describe
```

```python
#Reference: https://www.geeksforgeeks.org/twitter-sentiment-analysis-using-python/
def clean_text(text): 
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", text).lower().split())

def get_sentimentscore(cleaned_text): 
    sentiment_score = TextBlob(cleaned_text).sentiment.polarity
    return sentiment_score

def get_subjectivityscore(cleaned_text): 
    subjectivity_score = TextBlob(cleaned_text).sentiment.polarity 
    return subjectivity_score 

def categorize_sentiment(score): 
    if score > 0: 
        sentiment = "Positive"
    elif score < 0: 
        sentiment = "Negative"
    else: 
        sentiment = "Neutral"
    return sentiment

def categorize_subjectivity(score): 
    if score > 0: 
        subjectivity = "Subjective"
    elif score < 0: 
        subjectivity = "Not Subjective"
    else: 
        subjectivity = "Neutral"
    return subjectivity
    
```

```python
df["cleaned_text"] = df["text"].apply(clean_text)
```

```python
df["sentiment_score"] = df["cleaned_text"].apply(get_sentimentscore)
```

```python
df["sentiment"] = df["sentiment_score"].apply(categorize_sentiment)
```

```python
df["subjectivity_score"] = df["cleaned_text"].apply(get_subjectivityscore)
```

```python
df["subjectivity"] = df["subjectivity_score"].apply(categorize_subjectivity)
```
